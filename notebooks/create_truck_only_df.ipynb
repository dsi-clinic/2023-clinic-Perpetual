```python
import pandas as pd
```

subset distance matrix to trucks only

```python
location_df = pd.read_csv("../data/full_service_locations.csv")
#get list of index locations for all bike service locations
bike_index = location_df.loc[(location_df.loc[:, "pickup_type"]==
                              "Bike"), :].index.values.tolist()
```

```python
len(bike_index)
#13 locations that are bike only
```

```python
#get the column names for all bike service locations
columns_for_bikes = [str(i) for i in bike_index]
```

```python
distance_matrix = pd.read_csv("../data/indoor_outdoor_dist_matrix.csv")
```

```python
#drop all rows of index with bike service
distance_matrix = distance_matrix.drop(bike_index, axis=0)
#drop all columns with bike service
distance_matrix = distance_matrix.drop(columns_for_bikes, axis=1)
```

```python
distance_matrix.reset_index(inplace=True, drop=True)
distance_matrix
```

```python
distance_matrix.to_csv("../data/truck_distances.csv", index=False)
```

subset locations df to trucks only

```python
location_df = pd.read_csv("../data/full_service_locations.csv")
location_df = location_df.loc[(location_df.loc[:, "pickup_type"]!="Bike"), :]
location_df.reset_index(inplace=True, drop=True)
```

```python
# fill in missing values for pickup and dropoff values
location_df.loc[location_df.loc[:, "Name"]=="Vargas Cut and Catch (listed above)", "Daily_Pickup_Totes"] = 1.0
location_df.loc[location_df.loc[:, "Name"]=="Vargas Cut and Catch (listed above)", "Weekly_Dropoff_Totes"] = 1.0

location_df.loc[location_df.loc[:, "Name"]=="Whitecaps", "Daily_Pickup_Totes"] = 1.0
location_df.loc[location_df.loc[:, "Name"]=="Whitecaps", "Weekly_Dropoff_Totes"] = 1.0

location_df.loc[location_df.loc[:, "Name"]=="Rail Road Museum", "Daily_Pickup_Totes"] = 1.0
location_df.loc[location_df.loc[:, "Name"]=="Rail Road Museum", "Weekly_Dropoff_Totes"] = 1.0
```

```python
location_df.to_csv("../data/truck_service_locations.csv", index=False)
```
