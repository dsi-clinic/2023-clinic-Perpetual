# create_bike_only_dfs
This notebook will subset the original galveston indoor+outdoor location and distance matrix dataframes to bike service only. 

```python
import pandas as pd
```

### Subset Galveston indoor + outdoor locations df to trucks only

```python
#read in the galveston locations df:
#       this is the original full_df we made by concatonating 
#       moody gardens + indoor df + outdoor df together
location_df = pd.read_csv("../data/indoor_outdoor_pts_galv.csv")
```

```python
#subset to bike only rows
bike_only_df = location_df.loc[(location_df.loc[:, "pickup_type"]=="Bike"), :]
#reset index
bike_only_df.reset_index(inplace=True, drop=True)
```

```python
bike_only_df
```

```python
bike_only_df.to_csv("../data/bike_service_pts_galv.csv", index=False)
```

### Subset Galveston indoor + outdoor distance matrix to bikes only

```python
#import distance matrix for orignial galveston indoor+ outdoor dataset
distance_matrix = pd.read_csv("../data/indoor_outdoor_distances_galv.csv")
```

```python
#get list of bike location indices
bike_indices = location_df.loc[(location_df.loc[:, "pickup_type"]==
                              "Bike"), :].index.values.tolist()
```

```python
bike_indices
```

```python
#get the column names for all bike service locations
columns_for_bikes = [str(i) for i in bike_indices]
```

```python
#subset rows and columns with bike service indicies
bike_distances = distance_matrix.loc[bike_indices, columns_for_bikes]
```

```python
#reset index and columns
bike_distances.reset_index(inplace=True, drop=True)
bike_distances.columns = [str(i) for i in range(len(bike_indices))]

bike_distances
```

```python
bike_distances.to_csv("../data/bike_distances_galv.csv", index=False)
```
